```python
import ast
import warnings
import re
warnings.filterwarnings("ignore")
```

```python
def load_text_data():
    with open("../data/x.txt") as f:
        x_file = f.readlines()

    with open("../data/y.txt") as f:
        y_file = f.readlines()

    x_list = []
    y_list_temp = []

    for x_val, y_val in zip(x_file, y_file):
        x_list.append((x_val.split(", \n")[0]))
        y_list_temp.append((y_val.split(" \n")[0]))

    y_list = []
    for y in y_list_temp:
        x = y
        x = ast.literal_eval(x)
        y_list.append(x)

    x_list_clean = []
    y_list_clean = []

    for x in x_list:
        x = re.sub(r'[^\w\s]', '', x)
        x = x.lower()
        x_list_clean.append(x)

    for y in y_list:
        for i in range(len(y)):
            y[i] = re.sub(r'[^\w\s]', '', y[i])
            y[i] = y[i].lower()
        y_list_clean.append(y)

    return x_list_clean, y_list_clean

x_text, y_text = load_text_data()
```

```python
input_list = x_text

output_list = []

for input_str in input_list:
    words_coords = input_str.split()
    words = []
    coords = []
    item=0
    while item < len(words_coords):
        if words_coords[item].isdigit():
            coords.append(int(words_coords[item]))
            item=item+1
        else:
            words.append(words_coords[item])
            if(words_coords[item+1].isalpha()):
                try:
                    words[-1]=words[-1]+ '_' + words_coords[item+1]
                except:
                    print(input_list.index(input_str))


                item=item+2

            else:
                item=item+1
               
    result = ''
    for i in range(len(words)):
        result += words[i] + ' '
        result += ' '.join(map(str, coords[i*4:(i+1)*4]))
        result += ' '
    output_list.append(result)
```

```python
input_list = my_list
output_list=[]
for input_str in input_list:
# Split the input string into object strings
    object_strs = input_str.split()
    # print(object_strs)

    # Initialize empty lists for the object coordinates
    object_coords = []

    # Iterate over the object strings and extract the coordinates
    for i in range(1, len(object_strs), 5):
        x1, y1, x2, y2 = map(int, object_strs[i:i+4])
        object_coords.append((x1, y1, x2, y2))
    # print(object_coords)    


    output_coords = []
    for x1, y1, x2, y2 in object_coords:
        output_coords.append(( x1 + (x2-x1)/2, y1 + (y2-y1)/2))
    # print(output_coords)
    # Combine the object names and coordinates into the output string
    output_str = ''
    n=len(output_coords)
    # print(n)
    
    for i in range(n):
        
        name = object_strs[i*5]
        # print(name)
        

        coords_str = ' '.join(map(str, output_coords[i]))
        if i == 0:
            output_str += f"{name} {coords_str}"
        else:
            output_str += f" {name} {coords_str}"

    # Print the output string
    output_list.append(output_str)
output_list
```

```python
def get_location(coord):
    x = coord[1]
    y = coord[2]
    location_x = ""
    location_y = ""
    if x >= 0 and x < 33:
        location_x = "left"
    elif x >= 33 and x < 66:
        location_x = "middle"
    elif x >= 66 and x <= 100:
        location_x = "right"
    if y >= 0 and y < 33:
        location_y = "top"
    elif y >= 33 and y < 66:
        location_y = "middle"
    elif y >= 66 and y <= 100:
        location_y = "down"
    return location_y + " " + location_x

lst = output_list
final_output=[]
for coord_str in lst:
    coords = coord_str.split()

    obj_coords = []
    for i in range(0, len(coords), 3):
        obj_coords.append([coords[i], float(coords[i+1]), float(coords[i+2])])
    # print(obj_coords)
    locations = []

    for coord in obj_coords:
        x=get_location(coord)
        if(x=='middle middle'):
            x='center'
        locations.append(coord[0] + " location " + x)

    output_str = " ".join(locations)
    final_output.append(output_str)
final_output
```
